```python
site = []
title = []
for i in range(len(df['URL'])):    
    site.append(df['URL'][i])
    title.append(df['URL'][i].split("/")[3])    
    
df2 = pd.DataFrame({'site':site,'title':title})
    
```

```python
from collections import Counter 

temp = pd.DataFrame(Counter(df2['title']).items())
temp[1].unique()
hold = (temp[temp[1]==1])[0]
hold = hold.reset_index()
```

```python

valid_urls = []
for i in range(len(df2['title'])):
    for j in range(len(hold)):
        if(df2['title'][i] == hold[0][j]):
            valid_urls.append(df2['site'][i])            
        
del valid_urls[-1]        
```

```python
validurls_df = pd.DataFrame({"site": valid_urls})
validurls_df.to_csv('valid_url_hc.csv')
```

```python
del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['recipe'].isnull() == True].index)])    
mdf = mdf.reset_index()

del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['title'].isnull() == True].index)])    
mdf = mdf.reset_index()

del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['preptime'].isnull() == True].index)])    
mdf = mdf.reset_index()

del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['cooktime'].isnull() == True].index)])    
mdf = mdf.reset_index()

del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['totaltime'].isnull() == True].index)])    
mdf = mdf.reset_index()

del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['serving'].isnull() == True].index)])    
mdf = mdf.reset_index()

del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['course'].isnull() == True].index)])    
mdf = mdf.reset_index()

del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['cuisine'].isnull() == True].index)])    
mdf = mdf.reset_index()

del mdf['index']
mdf = mdf.drop(mdf.index[list(mdf[mdf['ingredients'].isnull() == True].index)])   
mdf = mdf.reset_index()

del mdf['index']
```

```python
mdf['cuisine'] = mdf['cuisine'].replace(['gujarat', 'gujarati', 'gujurati'],'gujarati')
mdf['cuisine'] = mdf['cuisine'].replace('indain','Indian')
mdf['cuisine'] = mdf['cuisine'].replace('indo chineese', 'indo chinese')
mdf['cuisine'] = mdf['cuisine'].replace(['interational', 'internatinal', 'international', 'internatonal', 'interntaional'],'international')
mdf['cuisine'] = mdf['cuisine'].replace(['maharashtra', 'maharashtrian'],'maharashtrian')
mdf['cuisine'] = mdf['cuisine'].replace(['rajasthan', 'rajasthani'],'rajasthani')
mdf['cuisine'] = mdf['cuisine'].replace(['north india', 'north indian'],'north indian')
mdf['cuisine'] = mdf['cuisine'].replace(['south india', 'south indiain', 'south indian'],'south indian')
mdf['cuisine'] = mdf['cuisine'].replace(['tamil', 'tamil nadu'],'tamil nadu')
```

```python
mdf['course'] = mdf['course'].replace(['sweet', 'sweets'],'sweet')
mdf['course'] = mdf['course'].replace(['Condiment', 'Condiments', 'pickle', 'sauce', 'powder', 'spice blend', 'dip'],'Condiment')
mdf['course'] = mdf['course'].replace(['sweet', 'sweets'],'sweets')
mdf['course'] = mdf['course'].replace('breakfast / snacks','snacks')
mdf['course'] = mdf['course'].replace(['sabji', 'sabzi'],'sabzi')
mdf['course'] = mdf['course'].replace(['tambli', 'raita'],'raita')
```
